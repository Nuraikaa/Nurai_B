```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vSQ21V8XmgRn3waBGFvkxovf5ZTQ7eoBPxwoXqNDwA_DRfo49mImvy_myMKGLfuyJzeuP07cYiDYqAM/pub?gid=1296240368&single=true&output=csv')
```

```python
sum = df.price.sum()
print("Общая стоимость бриллиантов в долларах: $", sum)

mean = df.price.mean()
print("Средняя стоимость бриллиантов в долларах: $", mean)
```

```python
df.describe()
```

```python
sns.set()
plt.figure(figsize=(20,6),dpi=300)
barplot = sns.barplot(data=df, x="cut",y="price" )
```

```python
prices = df['price']

mean_price = prices.mean()
median_price = prices.median()
min_price = prices.min()
max_price = prices.max()

print("Средняя цена: ", mean_price)
print("Медианная цена: ", median_price)
print("Минимальная цена: ", min_price)
print("Максимальная цена: ", max_price)

plt.hist(prices, bins=50)
plt.xlabel('Цена')
plt.ylabel('Количество')
plt.title('Распределение цен на алмазы')
plt.show()
```

```python
import matplotlib.pyplot as plt
carat = df.carat
clarity = df.clarity
plt.scatter(clarity, carat)
plt.show()

#Этот график представляет собой точечную диаграмму, показывающую чистоту бриллианта в сравнении с размером бриллианта в каратах
```

```python
clarityindexes = df["clarity"].value_counts().index.tolist()
claritycount = df["clarity"].value_counts().values.tolist()

print(clarityindexes)
print(claritycount)
plt.bar(clarityindexes, claritycount)
plt.show()

#вторая визуализация представляет собой гистограмму для визуализации количества бриллиантов в каждом типе чистоты
```
